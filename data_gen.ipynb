```python
import pandas as pd
from umap import UMAP
from sklearn.pipeline import make_pipeline 
from embetter.text import SentenceEncoder
```

```python
model_name = 'all-MiniLM-L6-v2'
text_emb_pipeline = make_pipeline(
    SentenceEncoder(model_name),
    UMAP()
)
```

```python
mm = pd.read_csv('../datasets/eliot_middlemarch.csv')
sentences = mm.sents.to_list()
```

```python
X_tfm = text_emb_pipeline.fit_transform(sentences)
```

```python
df = pd.DataFrame({"text":sentences, "chapter":mm.chapter})
df['x'] = X_tfm[:, 0]
df['y'] = X_tfm[:, 1]
```

```python
df['sent_size'] = df['text'].apply(lambda x: len(x.split()))
df 
```

```python
df.to_csv('mm_embedding.csv')
```
